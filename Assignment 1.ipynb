```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line plot")
```

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.plot(days,sales_1,'go-', linewidth = 1, markersize=10)
plt.plot(days,sales_2,'bo-', linewidth = 1, markersize=10)
plt.title("Sales Plot")
plt.xlabel("days")
plt.ylabel("sales")
```

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```
